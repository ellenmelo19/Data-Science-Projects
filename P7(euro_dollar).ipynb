# Visualização de dados de Storytelling sobre taxas de câmbio

## Dados

```python
import pandas as pd

taxas_de_cambio = pd.read_csv('euro-daily-hist_1999_2020.csv')
taxas_de_cambio.head()
```

```python
taxas_de_cambio.tail()
```

```python
taxas_de_cambio.info()
```

## Limpando os dados

```python
taxas_de_cambio.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
taxas_de_cambio['Time'] = pd.to_datetime(taxas_de_cambio['Time'])
taxas_de_cambio.sort_values('Time', inplace=True)
taxas_de_cambio.reset_index(drop=True, inplace=True)
taxas_de_cambio.head()
```

```python
euro_p_dollar = taxas_de_cambio[['Time', 'US_dollar']].copy()
euro_p_dollar['US_dollar'].value_counts() # 62 '-' characters
```

```python
euro_p_dollar = euro_p_dollar[euro_p_dollar['US_dollar'] != '-']
euro_p_dollar['US_dollar'] = euro_p_dollar['US_dollar'].astype(float)
euro_p_dollar.info()
```

### Média

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_p_dollar['Time'], euro_p_dollar['US_dollar'])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_p_dollar['Time'], euro_p_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, media in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_p_dollar['Time'],
             euro_p_dollar['US_dollar'].rolling(media).mean())
    plt.title('Rolling Window:' + str(media), weight='bold')
    
plt.tight_layout() # Ajusta automaticamente o preenchimento
plt.show()
```

```python
euro_p_dollar['media'] = euro_p_dollar['US_dollar'].rolling(30).mean()
euro_p_dollar
```

## Sugestão

Algumas ideias de como trabalhar com nossos dados:

* Mostrar como a cotação euro-dólar mudou durante a pandemia de coronavírus. Podemos usar os dados de 2020 e os dados de 2016-2019 como linha de base.

* Mostrar como a taxa euro-dólar mudou durante a crise financeira de 2007-2008 (os dados de 2016 e 2009 podem ser usados para comparação). 

* Comparar como a taxa euro-dólar mudou sob os três últimos presidentes dos EUA (George W. Bush (2001-2009), Barack Obama (2009-2017) e Donald Trump (2017-2021)).

## RP de visualização de dados 

### Crise Financeira

```python
crise = euro_p_dollar.copy(
                   )[(euro_p_dollar['Time'].dt.year >= 2006
                   ) & (euro_p_dollar['Time'].dt.year <= 2009)]
crise_7_8 = euro_p_dollar.copy(
                   )[(euro_p_dollar.Time.dt.year >= 2007
                   ) & (euro_p_dollar.Time.dt.year <= 2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

fig,ax = plt.subplots(figsize=(8,3))
ax.plot(crise['Time'],
        crise['rolling_mean'],
        linewidth=1, color='#A6D785')

### Destacando o período 2007-2008
ax.plot(crise_7_8['Time'],
        crise_7_8['rolling_mean'],
        linewidth=3, color='#e23d28')

### Destacando o pico da crise
ax.axvspan(xmin=733112.0, xmax=733302.0, ymin=0.09,
           alpha=0.3, color='grey')
    

ax.set_xticklabels([])
ax.set_yticklabels([])

x = 732272.0
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365

y = 1.193   
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(732172.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.1

## Adicionando um título e uma legenda    
ax.text(732172.0, 1.67, "A taxa Euro-USD atingiu um pico de 1,59 durante a crise financeira de 2007-2008",
        weight='bold')
ax.text(732172.0, 1.63, 'Taxas de câmbio Euro-USD entre 2006 e 2010',
        size=12)


ax.text(732172.0, 1.07, '©DATAQUEST' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

### Adicione um pouco de transparência à grade
ax.grid(alpha=0.5)

plt.show()
```

### 3 Últimos Presidentes

```python
bush_obama_trump = euro_p_dollar.copy(
                   )[(euro_p_dollar['Time'].dt.year >= 2001) & (euro_p_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

O que faremos:

* Primeiro construiremos três dos gráficos em uma grade de 2 por 3.
* construiremos apenas o gráfico inferior de uma grade 2 por 1.
* As duas grades são mescladas e terminamos com três gráficos na linha superior e um gráfico na linha inferior.

```python
style.use('fivethirtyeight')


plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]


for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    
    

### Bush
ax1.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)
ax1.text(731516.0, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(731216.0, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)


### Obama
ax2.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'],
                   alpha=0.3)
ax2.text(734288.0, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(734138.0, 1.8, '(2009-2017)', weight='bold',
         alpha=0.3)


### Trump
ax3.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)
ax3.text(736855.0, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(736745.0, 1.8, '(2017-2021)', weight='bold',
         alpha=0.3)

### Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adicionando Título e Subtítulo
ax1.text(730016.0, 2.35, 'Taxa EURO-USD média de 1,22 sob os últimos três presidentes dos EUA',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''Taxas de câmbio EURO-USD sob George W. Bush (2001 - 2009), Barack Obama (2009-2017),
e Donald Trump (2017-2021)''',
        fontsize=16)


ax4.text(729916.0, 0.65, '©DATAQUEST' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```
